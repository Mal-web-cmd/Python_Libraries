Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Wholesale customers data.csv")
x = dataset.iloc[:,[2,4]].values
```

```python
print(x)
```

Optimisation

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters= i, init= 'k-means++', random_state = 21)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('WCSS via Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS Value')

plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_means = kmeans.fit_predict(x)
```

```python
print(y_means)
```

K means clustering

```python
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'yellow', label = 'Cluster 3')
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s = 100, c = 'green', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'black', label = 'Centroids')
plt.title('Wholesale customers data')
plt.xlabel('Detergents_Paper')
plt.ylabel('Delicassen')
plt.legend()
plt.show()
```

optimisation

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Delicassen')
plt.ylabel('Euclidean distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 6, linkage = 'ward')
yhc = hc.fit_predict(x)
```

hierarchical clustering

```python
plt.scatter(x[yhc == 0, 0], x[yhc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[yhc == 1, 0], x[yhc == 1, 1], s = 100, c = 'yellow', label = 'Cluster 2')
plt.scatter(x[yhc == 2, 0], x[yhc == 2, 1], s = 100, c = 'blue', label = 'Cluster 3')
plt.scatter(x[yhc == 3, 0], x[yhc == 3, 1], s = 100, c = 'green', label = 'Cluster 4')
plt.scatter(x[yhc == 4, 0], x[yhc == 4, 1], s = 100, c = 'pink', label = 'Cluster 5')
plt.scatter(x[yhc == 5, 0], x[yhc == 5, 1], s = 100, c = 'black', label = 'Cluster 6')

plt.title('Wholesale Customers data')
plt.xlabel('Detergents_Paper')
plt.ylabel('Delicassen')

plt.legend()
plt.show()
```

DBSCAN

```python

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 5, min_samples = 5)
y_dbscan = dbscan.fit_predict(x)
```
